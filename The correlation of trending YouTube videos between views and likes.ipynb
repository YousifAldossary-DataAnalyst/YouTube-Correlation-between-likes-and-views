```python
## triming down the number of columns to find the correlation.

cor_list = ['views', 'likes', 'dislikes', 'comment_count']
hm_data = DE_yt[cor_list].corr() 
display(hm_data)
```

As always in correlations: <Br>
     1 = strong poisitn correlation.<br>
     0 = no correlation<br>
     -1 = strong negative correlation <br>
<br>
This is pearson correlation. there is strong correlation between number of like and views, while high likely correlation with number of comments count with views.
